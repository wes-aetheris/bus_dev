# Sensor Analysis Notebook

This notebook provides analysis and visualization for drone sensor data, including health trends, environmental factors, and physics-based degradation modeling.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

## Load Sensor Data

We load the actual sample sensor data from the project.

```python
df = pd.read_csv('../data/sample_sensor_data.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

## Basic Statistics

Get a quick overview of sensor health and environmental factors.

```python
df.describe()
```

## Plot Sensor Health Over Time

Visualize health trends for each sensor.

```python
plt.figure(figsize=(12,7))
for sensor in df['sensor_id'].unique():
    plt.plot(df[df['sensor_id']==sensor]['timestamp'], df[df['sensor_id']==sensor]['health'], label=sensor)
plt.xlabel('Timestamp')
plt.ylabel('Health (%)')
plt.title('Sensor Health Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
